```python
import pandas as pd
import numpy as np
diabetes = pd.read_csv('pima_diabetes.csv', delimiter='	')
pd.options.display.max_columns = 99
pd.options.display.width = 150
print(diabetes.head(10))
```

```python
columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
diabetes = pd.read_csv('pima_diabetes.csv', delimiter='	', na_values='  ', names=columns)
print(diabetes.head(10))
```

```python
#Since, there are can be a number of other factors that can be used for identifying diabetes apart from the features 
#listed in this dataset like increased thirst, increased hunger (especially after eating), dry mouth etc..
#We will use Gaussian Naive Bayes Classifier, for modelling probabilistic relationship between the attributes and 
#the target variable.
#At first we will remove the null values.
```

```python
print(diabetes.isnull().sum())
```

```python
#Replacing the null values in the columns with their median.
x = diabetes.isnull().sum()
df_missingvals = diabetes[x[(x>0)].index]
diabetes[df_missingvals.columns] = df_missingvals.fillna(df_missingvals.median())
print(diabetes.isnull().sum())
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,8))
ax1 = fig.add_subplot(4,2,1)
ax2 = fig.add_subplot(4,2,2)
ax3 = fig.add_subplot(4,2,3)
ax4 = fig.add_subplot(4,2,4)
ax5 = fig.add_subplot(4,2,5)
ax6 = fig.add_subplot(4,2,6)
ax7 = fig.add_subplot(4,2,7)
ax8 = fig.add_subplot(4,2,8)
ax3.boxplot(diabetes['BloodPressure'])
ax2.boxplot(diabetes['Glucose'])
ax1.boxplot(diabetes['Pregnancies'])
ax4.boxplot(diabetes['SkinThickness'])
ax5.boxplot(diabetes['Insulin'])
ax7.boxplot(diabetes['DiabetesPedigreeFunction'])
ax6.boxplot(diabetes['BMI'])
ax8.boxplot(diabetes['Age'])
plt.show()
```

```python
#We can see a lot of outliers in the dataset,
#Also we see a number of 0 values in columns where it looks odd. Like..
#A living person cannot have diastolic blood pressure of zero so, the readings of the data set seems wrong.
#Even after fasting glucose level would not be as low as zero. Therefor zero is an invalid reading.
#For normal people skin fold thickness canâ€™t be less than 10 mm better yet zero.
#BMI should not be 0 or close to zero unless the person is really underweight which could be life threatening. 
#In a rare situation a person can have zero insulin but by observing the data, we can find more number of 0 values present in it.
#We will replace the zero values with the mean of each column.
```

```python
columns = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
diabetes[columns] = diabetes[columns].replace(0, diabetes[columns].mean())
```

```python
#Checking collinearity within the features.
corrmat = diabetes.corr()
price_corrs = corrmat['Outcome'].abs().sort_values()
import seaborn as sns
import matplotlib.pyplot as plt
fig,axes = plt.subplots(figsize=(10,8))
sns.heatmap(corrmat.abs(), annot=True)
plt.show()
```

```python
#seems like SkinThickness and BMI are correlated. also, Age and Pregnancies are correlated.
#Gaussian NB assumes the feature variables to be independent of each other. So we'll have to remove any one from each pair.
```

```python
#scaling the data.
minimum = diabetes.min(axis=0)
maximum = diabetes.max(axis=0)
diabetes = (diabetes - minimum)/(maximum - minimum)
variance = diabetes.drop('Outcome', axis=1).var()
print(variance)
```

```python
#Removing 'SkinThickness' and 'Age' because of lower variance.
diabetes = diabetes.drop(['SkinThickness','Age'], axis=1)
```

```python
#Preparing the training and test dataset.
shuffled_index = np.random.permutation(diabetes.index)
df = diabetes.reindex(shuffled_index)
train_size = int(0.75*df.shape[0])
train_df = df.iloc[:train_size].copy()
test_df = df.iloc[train_size:].copy()
```

```python
#Grouping the training data on the basis of Outcome.
train_cols = diabetes.columns.drop('Outcome')
grouped = train_df.pivot_table(index = 'Outcome', values = train_cols, aggfunc = [np.mean, np.std])
print('Standard deviation of each feature in each group:')
print(grouped['std'])
print('Mean of each feature in each group:')
print(grouped['mean'])
```

```python
print(grouped.index)
```

```python
#Calculating prior probability for each class.
prob_yes = len(train_df[train_df['Outcome']==1]) / len(train_df)
prob_no = len(train_df[train_df['Outcome']==0]) / len(train_df)
print('prob_yes: ', str(prob_yes), 'prob_no: ', str(prob_no))
```

```python
#Function to calculate the probability of a row belonging to each class.
import math
def predict(row):
    #Given a row, the post_prob list will contain the Posterior Probabilities of the row for each class.
    post_prob = list()
    for classes in grouped.index:
        cond_prob = 1
        for i in range(len(row)-1):
            attr = train_cols[i]
            mean = grouped.loc[classes, ('mean',attr)]
            stdev = grouped.loc[classes, ('std',attr)]
            exponent = math.exp(-(math.pow(row[i] - mean,2)/(2*math.pow(stdev,2))))
            cond_prob *= (1 / (math.sqrt(2*math.pi) * stdev)) * exponent
        if(classes == 0.0):
            cond_prob *= prob_no
        else:
            cond_prob *= prob_yes
        post_prob.append(cond_prob)
    return post_prob
```

```python
#Making predictions on the test dataset.
predictions = []
for i in range(len(test_df)):
    pred_no, pred_yes = predict(test_df.iloc[i])
    if(pred_no >= pred_yes):
        pred = 0
    else:
        pred = 1
    predictions.append(pred)
```

```python
#Calculating accuracy of the model using roc_auc_score.
from sklearn.metrics import roc_auc_score
score = roc_auc_score(test_df['Outcome'],predictions)
print(score)
```
